```python
# Conver the date into a suitable format
data["date"] = pd.to_datetime(data["date"])
```

```python
data["target"] = (data["result"] == "W").astype("int")
```

```python
data ["venue_code"] = data["venue"].astype("category").cat.codes
```

```python
# Using the Machine Learning model

from sklearn.ensemble import RandomForestRegressor
```

```python
# Creating an instance of the model
rf = RandomForestRegressor(n_estimators = 50, random_state = 1, min_samples_split = 20)
```

```python
train = data[data["date"] < '2022-01-01']
```

```python
test = data[data["date"] > '2022-01-01']
```

```python
predictors = ["venue_code", "opp_code", "day_code", "hour"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
y = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
binary_preds = (y>=0.5).astype(int)
```

```python
error = accuracy_score(test["target"], binary_preds)
```

```python
grouped_matches = data.groupby("team")
```

```python
group = grouped_matches.get_group("Manchester City").sort_values("date")
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = data.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], binary_preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```

```python
error
```

```python
combined = combined.merge(matches_rolling[["date", "team", "opponent", "result"]], left_index=True, right_index=True)
```

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)
```

```python
combined["new_team"] = combined["team"].map(mapping)
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
merged
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a line plot for the rolling averages
rolling_fig = go.Figure()
for col in new_cols:
    rolling_fig.add_trace(go.Scatter(x=matches_rolling['date'], y=matches_rolling[col], mode='lines', name=col))

rolling_fig.update_layout(title='Rolling Averages',
                          xaxis_title='Date',
                          yaxis_title='Value')

# Create a scatter plot for the match predictions
scatter_fig = go.Figure()
scatter_fig.add_trace(go.Scatter(x=combined['date'], y=combined['team'], mode='markers', 
                                marker=dict(color=combined['predicted'], symbol=combined['actual']),
                                text=combined['opponent'], hovertemplate='Team: %{y}<br>Opponent: %{text}<extra></extra>'))
scatter_fig.update_layout(title='Match Predictions',
                          xaxis_title='Date',
                          yaxis_title='Team')

# Create subplots
fig = make_subplots(rows=2, cols=1, shared_xaxes=True)

# Add rolling averages subplot
for trace in rolling_fig.data:
    fig.add_trace(trace, row=1, col=1)

# Add match predictions subplot
for trace in scatter_fig.data:
    fig.add_trace(trace, row=2, col=1)

# Set subplot titles and axis labels
fig.update_layout(title='Interactive Match Analysis')

# Update layout for subplot arrangement
fig.update_layout(height=600, width=800, showlegend=False)

# Update annotations for subplot titles
fig.update_annotations(rolling_fig.layout.annotations + scatter_fig.layout.annotations)

# Show the combined plot
fig.show()
```
